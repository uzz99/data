```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/pok/Pokemon.csv')
df.head()
```

```python
df[df['Legendary'] == True].HP.mean() -df[df['Legendary'] == False].HP.mean()
```

```python
df['Type 2'].value_counts().index[0]
```

```python
t1 = df['Type 1'].value_counts().index[0]
df[df['Type 1'] == t1]['Attack'].mean() / df[df['Type 1'] == t1]['Defense'].mean()
```

```python
df[df['Legendary']==True]['Generation'].value_counts().index[0]
```

```python
res = df[[ 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].corr().unstack().reset_index().rename(columns = {0:'corr'})
res[res['corr']!=1].sort_values('corr',ascending=False).iloc[0]

#1은 자기 자신과의 상관계수이므로 제외
```

```python
df.sort_values(['Generation','Attack']).groupby('Generation').head(3)['Attack'].mean()
# Generation 별로 3개의 데이터를 선택하기 위해 groupby, groupby를 사용안하면 전체 데이터에서 상위 3개만 선택하게 됨
```

```python
df.sort_values(['Generation','Attack'],ascending=False).groupby('Generation').head(5)['Attack'].mean()
```
