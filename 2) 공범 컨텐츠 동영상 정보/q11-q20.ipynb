```python
import pandas as pd

df=pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/channelInfo.csv')
df_video =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/videoInfo.csv')
```

```python
df['ct']=pd.to_datetime(df['ct'])
df_video['ct']=pd.to_datetime(df_video['ct'])
df_video['videoname'].value_counts()
```

```python
df_video.sort_values(['videoname','ct']).drop_duplicates('videoname', keep='last')[['viewcnt','videoname','ct']].reset_index(drop=True)
```
