## 기온 강수량 데이터



```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/weather/weather2.csv")
df.head()
```

```python
df['time'] = pd.to_datetime(df['time'])

summer= df[df['time'].dt.month.isin([6,7,8])]

len(summer[summer['이화동기온'] > summer['수영동기온']])
```

```python
print(df[df['이화동강수'].max() == df['이화동강수']].time.values)
print(df[df['수영동강수'].max() == df['수영동강수']].time.values)
```

## 서비스 이탈예측 데이터



```python
import pandas as pd
#데이터 로드
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churn/train.csv")
df.head()
```

```python
df[df['Gender'] == 'Male'].groupby('Geography').sum().sort_values('Exited',ascending=False).head(1).reset_index()[['Geography','Exited']]
```

```python
round(df[(df['HasCrCard'] == 1) & (df['IsActiveMember'] == 1)]['Age'].mean(),4)
```

```python
round(df[df['Balance'] >= df['Balance'].median()]['CreditScore'].std(),3)
```

### 성인 건강검진 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/smoke/train.csv")
df.head()
```

```python
df['혈압차'] = df['수축기혈압'] -df['이완기혈압']
df.groupby('연령대코드(5세단위)')['혈압차'].var().sort_values().index[-5]
```

```python
df['WHtR'] = df['허리둘레']/df['신장(5Cm단위)']
res = df[df['WHtR'] >= 0.58]['성별코드'].value_counts()

res['M']/res['F']
```
