```python
import pandas as pd
df =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/body/body.csv')
df.head()
```

```python
(df['수축기혈압(최고) : mmHg'] - df['이완기혈압(최저) : mmHg']).mean()
```

```python
df[(df['측정나이'] < 60) & (df['측정나이'] >= 50)]['신장 : cm'].mean()
```

```python
df['연령대'] = df['측정나이'] // 10 * 10
res = df['연령대'].value_counts().to_frame()
res.rename(columns={'count':'인원수'})
```

```python
df.groupby(['연령대', '등급']).size().reset_index().rename(columns={0:'size'})
```

```python
male = df[df['측정회원성별'] == 'M']
male
male[male['등급'] == 'D']['체지방율 : %'].mean() - male[male['등급'] == 'A']['체지방율 : %'].mean()
```

```python
female = df[df['측정회원성별'] == 'F']
female
female[female['등급'] == 'D']['체중 : kg'].mean() - female[female['등급'] == 'A']['체중 : kg'].mean()
```

```python
df['bmi'] = df['체중 : kg'] / (df['신장 : cm']/100)**2
df[df['측정회원성별'] == 'M']['bmi'].mean()
```

```python
df[df['bmi'] < df ['체지방율 : %']]['체중 : kg'].mean()
```

```python
df[df['측정회원성별'] == 'M']['악력D : kg'].mean() - df[df['측정회원성별'] == 'F']['악력D : kg'].mean()
```
