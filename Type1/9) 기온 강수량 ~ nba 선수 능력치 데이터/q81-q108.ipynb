### 성인 건강검진 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/smoke/train.csv")
df.head()
```

```python
df['혈압차'] = df['수축기혈압'] -df['이완기혈압']
df.groupby('연령대코드(5세단위)')['혈압차'].var().sort_values().index[-5]
```

```python
df['WHtR'] = df['허리둘레']/df['신장(5Cm단위)']
res = df[df['WHtR'] >= 0.58]['성별코드'].value_counts()

res['M']/res['F']
```

### 의료 비용 예측 데이터

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/train.csv")
df.head()
```

```python
smoker = df[df['smoker'] == 'yes']
non = df[df['smoker']=='no']

high_yes = smoker['charges'].quantile(0.9)
high_no = non['charges'].quantile(0.9)

mean_yes = df[(df.smoker =='yes') &(df.charges >=high_yes)].charges.mean()
mean_no = df[(df.smoker =='no') &(df.charges >=high_no)].charges.mean()

mean_yes - mean_no
```
