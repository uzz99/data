```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/consum/Tetuan%20City%20power%20consumption.csv')
df.head()
```

```python
df['DateTime'] = pd.to_datetime(df['DateTime'])
df['DateTime'].dt.month.value_counts().sort_index()
```

```python
res = df[df['DateTime'].dt.month == 3]

#시간대별
res.groupby(res.DateTime.dt.hour)['Temperature'].mean().min()
```

```python
res = df[df['DateTime'].dt.month == 3]

#시간대별
res.groupby(res.DateTime.dt.hour)['Temperature'].mean().max()
```

```python
# df = df.rename(columns={'Zone 2  Power Consumption': 'Zone 2 Power Consumption'})
df[df['Zone 1 Power Consumption'] > df['Zone 2 Power Consumption']].Humidity.mean()
```

```python
df.iloc[:,-3:].corr()
```

```python
def tem(x):
    if x<10:
        return 'A'
    elif x<20:
        return 'B'
    elif x<30:
        return 'C'
    else:
        return 'D'

df.Temperature.apply(tem).value_counts()
```
