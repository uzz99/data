```python
import pandas as pd

#시험에서는
# data = pd.read_csv('')# 원본데이터
# X = data.drop(columns=['Exited'])  # 독립 변수(x에는 exited가 포함되면 x)
# y = data['Exited']  # 종속 변수(exited만 포함되어야)
# from sklearn.model_selection import train_test_split
# x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)



x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_test.csv")

```

```python
x_train.info()
y_train.info()

display(x_train.head())
display(y_train.head())

print(x_train.isna().sum())
print()
print(x_test.isna().sum())
print()
print(y_train.isna().sum())
```

```python
# 예측과 무관한 열 제거
x_train_drop=x_train.drop(columns = ['CustomerId','Surname']) # 고객식별 정보 및 이름
x_test_drop=x_test.drop(columns = ['CustomerId','Surname'])

display(x_train_drop)
display(x_test_drop)
```

```python
from sklearn.model_selection import train_test_split#데이터 분리
from sklearn.ensemble import RandomForestClassifier#랜포

# 범주형 데이터 처리, Gender(Male, Female)가 → Gender_Male, Gender_Female의 두 개 열로 변환
# 데이터 전처리(원핫인코딩)
x_train_dummies = pd.get_dummies(x_train_drop)
x_test_dummies = pd.get_dummies(x_test_drop)
```

```python
x_train_dummies
```

```python
x_test_dummies
```

```python
y = y_train.Exited
y
```

```python
# train과 컬럼 순서 동일하게 하기 (더미화 하면서 순서대로 정렬을 이미 하기 때문에 오류가 난다면 해당 컬럼이 누락된것)
#학습 데이터와 테스트 데이터의 컬럼 순서를 맞춤
x_test_dummies = x_test_dummies[x_train_dummies.columns]
# print(help(train_test_split))
display(x_test_dummies)

#데이터 분리
X_train, X_validation, Y_train, Y_validation = train_test_split(x_train_dummies, y, test_size=0.3, random_state=42)

# 모델 학습
rf = RandomForestClassifier(random_state=23)
rf.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score , f1_score, recall_score, roc_auc_score ,precision_score

# 모델 예측 및 평가
predict_train_label = rf.predict(X_train)
predict_train_proba = rf.predict_proba(X_train)[:,1]

predict_validation_label = rf.predict(X_validation)
predict_validation_prob = rf.predict_proba(X_validation)[:,1]
```

```python
from sklearn.metrics import accuracy_score
print('train accuracy :', accuracy_score(Y_train,predict_train_label))
print('validation accuracy :', accuracy_score(Y_validation,predict_validation_label))
```

```python
predict_test_label = rf.predict(x_test_dummies)
prdiect_test_proba = rf.predict_proba(x_test_dummies)[:,1]

pd.DataFrame({'CustomerId':x_test.CustomerId, 'Exited': prdiect_test_proba}).to_csv('result.csv', index=False)
```
