데이터 설명 : 비행탑승 경험 만족도 (satisfaction 컬럼 : ‘neutral or dissatisfied’ or satisfied ) (83123, 24) shape

```python
import pandas as pd
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/x_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/x_test.csv")
```

```python
display(x_train.head())
display(y_train.head())

print(x_train.info())
print(x_test.info())
```

```python
print(x_train.isna().sum())
print(x_test.isna().sum())
```

```python
x_train['Arrival Delay in Minutes'] = x_train['Arrival Delay in Minutes'].fillna(x_train['Arrival Delay in Minutes'].mean())
x_test['Arrival Delay in Minutes'] = x_test['Arrival Delay in Minutes'].fillna(x_test['Arrival Delay in Minutes'].mean())
```

```python
print(x_train.isna().sum())
print(x_test.isna().sum())
```

```python
x_train_drop = x_train.drop(columns = ['ID', 'id'])
x_test_drop = x_test.drop(columns = ['ID', 'id'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

x_train_dum = pd.get_dummies(x_train_drop)
x_test_dum = pd.get_dummies(x_test_drop)
x_test_dum = x_test_dum[x_train_dum.columns]

y = y_train['satisfaction']
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_train_dum, y, test_size=0.3, random_state=23)
rf = RandomForestClassifier(random_state=23)
rf.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score, roc_auc_score, precision_score

predict_train = rf.predict(x_train)
predict_train_proba = rf.predict_proba(x_train)[:,0]

predict_val = rf.predict(x_val)
predict_val_proba = rf.predict_proba(x_val)[:,0]
```

```python

print('--accuracy_score--')
print(accuracy_score(y_train, predict_train))
print(accuracy_score(y_val, predict_val))
print()
print('--f1_score--')
print(f1_score(y_train, predict_train, pos_label = 'neutral or dissatisfied'))
print(f1_score(y_val, predict_val, pos_label = 'neutral or dissatisfied'))
print()
print('--recall_score--')
print(recall_score(y_train, predict_train, pos_label = 'neutral or dissatisfied'))
print(recall_score(y_val, predict_val, pos_label = 'neutral or dissatisfied'))
print()
print('--roc_auc_score--')
print(roc_auc_score(y_train, predict_train_proba))
print(roc_auc_score(y_val, predict_val_proba))
print()
print('--precision_score--')
print(precision_score(y_train, predict_train, pos_label = 'neutral or dissatisfied'))
print(precision_score(y_val, predict_val, pos_label = 'neutral or dissatisfied'))

# pos_label -> 라벨이 문자로 표현된 경우 or 관심 클래스를 지정해야 하는 경우
```

```python
# 클래스 확인
# sklearn은 기본적으로 많은 값을 0번 클래스에 배치
print(y_train.value_counts())
```
